```python
# import required libraries
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from tpot import TPOTClassifier
```

```python
#load data
cancer = load_breast_cancer()
X = cancer.data
Y = cancer.target
feature_names= cancer.feature_names
```

```python
# load to dataframe
df = pd.DataFrame(X, columns= feature_names)
df['target']= Y
df.head()
```

```python
#split data
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2, random_state=42)
#scalar = StandardScaler()
#X_train = scalar.fit_transform(X_train)
#X_test = scalar.transform(X_test)
```

```python
# intiate model
tpot = TPOTClassifier(verbosity=2, generations=10, population_size=20,config_dict='TPOT sparse', random_state=42)

# Fit the model on the training data
tpot.fit(X_train, Y_train)
```

```python
# Access the evaluated individuals
evaluated_pipelines = tpot.evaluated_individuals_

# Print the evaluated pipelines and their scores
for pipeline, score in evaluated_pipelines.items():
    print(f"Pipeline: {pipeline}\nScore: {score}\n")
```

```python
# Extract the best model
best_model = tpot.fitted_pipeline_

# Print the best model
print(f" Best model is: {best_model}")

# Evaluate the best model on the testing data
accuracy = best_model.score(X_test, Y_test)
print(f"Accuracy of the best model: {accuracy}")


# Print the accuracy
print("Test accuracy: %.3f" % tpot.score(X_test, Y_test))
```

```python
# Export the best pipeline
tpot.export('tpot_breast_cancer_pipeline.py')
```
